## 挑戰題一：買相機

題目說明：財團法人大兔兔文教基金會要添購攝影設備

他們預計在未來的四個月內分別添購四項價格為

100k，150k，230k，80k的設備

請以折線圖畫出累積消費額，並以圓餅圖畫出各器材所佔比例

```python
list_price = [100000 , 150000 , 230000 , 80000]
list_item = ["A", "B", "C", "D"]
```

```python
#累積消費
list_accumulated_price = []

accumulated_price = 0
for price in list_price:
    accumulated_price = accumulated_price + price
    list_accumulated_price.append(accumulated_price)
```

```python
plt.xticks(list(range(0,4)), list_item)
plt.plot(list_accumulated_price, marker="o")
```

```python
plt.pie(list_price, labels=list_item, autopct='%1.1f%%')
```
